```python
# tuple data filtering use case

t1 = (10,50,40,50,60,80,90,("ridoy","python","inception"),80,70,80,90,50,90,80,10)
t2 = []
t3 = []
for i in t1:
    if type(i) == int:
        t2.append(i)
    if type(i) == tuple:
        for j in i:
            if type(j) == str:
                t3.append(j)


t2 = tuple(t2)
t3 = tuple(t3)

print(t2)
print(t3)
```
