```python
# Dictionaries function 
d = {1 : "Ashik",
    2 : "Sagor",
    3 : "Habib"}
print(d)
```

```python
# Dictionaries function indexing 
d1 = {"Ashik" : 541,
     "Sagor" : 540,
     "Habib" : 510}
print(d1)
d1["Ashik"]
```

```python
# items() function bebohar korle jeto gula item asche sob items er list diye dibe

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.items()
print(d1)
```

```python
# Dictionaries or item() diye for loop run

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.items()
for i, j in d1.items():
    print(f"key = {i}, valu = {j}")
```

```python
# keys function diye for loop run

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.keys()
for i in d1.keys():
    print(i)
```

```python
# values function diye for loop run

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.values()
for i in d1.values():
    print(i)
```

```python
# pop function kono kichu valu name dore delete korte caile pop function use kore

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.pop("Sagor")
print(d1)
```

```python
# popitem function sob somoy last er item take delete kore

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.popitem()
print(d1)
```

```python
# del function use case

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
del d1
print(d1)
```

```python
# del item use case
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
del d1["Ashik"]
print(d1)
```

```python
# update function holo 2ta dictionari ke 1ta dictionari mane old dictionari te print kore

d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d2 = {"a" : 444,
     "b" : 443,
     "c" : 442,
     "d" : 441,
     "e" : 440,
     "f" : 439,
     "g" : 438}
d1.update(d2)
print(d1)
```

```python
# get function holo jodi indexing kori tahole valu thake indexing hobe are na thakle kono error dibe na
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.get("x")

"""
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.["x"]
"""
```

```python
# clear function dictionaries e bebohar korle dictionari khali ba blank hoye jabe
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.clear()
print(d1)
```

```python
# copy function diye dictionari copy kora jay
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d2 = d1.copy()
print(d1)
print(d2)
```

```python
# fromkeys function er vitor kichu dile sob gula keys hoye jay
d1.fromkeys("ridoy")
```

```python
# onek keys and valus chaile fromkeys function use korte pari
a = {}.fromkeys(range(1,100), "inception bd")
#print(a)
```

```python
# setdefault function holo key na thkle automatic key add hoye jabe 
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.setdefault("ridoy")
print(d1)
```

```python
# setdefault function holo key and valu na thkle automatic key and valu add hoye jabe 
d1 = {"Ashik" : 541,
    "Sagor" : 540,
    "Habib" : 510}
d1.setdefault("anamul", 600)
print(d1)
```

```python
# nested Dictionaries function use case

d = {"student_name" : ["ridoy", "bappy", "debasish", "mamun"],
    "course_name" : "python",
    "course_du" : 6,
    "topics" : {"python_basic" : "if, else, loop",
               "opps" : "1, 2, 3"}
}
print(d)
```
