```python
# Resut Mark Sheet

a = int(input("please enter your mark: "))

if a>= 80:
    print(f"congratulations, your mark is {a}, and resut is A+")
elif a>= 70:
    print(f"congratulations, your mark is {a}, and resut is A")
elif a>= 60:
    print(f"congratulations, your mark is {a}, and resut is A-")
else:
    print(f"Sorry, your mark is {a}, and resut is F")
```
