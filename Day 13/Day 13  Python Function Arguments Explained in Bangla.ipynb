```python
# Required Arguments

def average(x,y):
    result = (x+y)/2
    return result

average(30,40)
```

```python
# keyword arguments
def average(x,y):
    result = (x+y)/2
    return result
average(50,60)
average(x = 50, y = 90)
average(y = 90, x = 50)
```

```python
# keyword arguments
def my_info(name, age):
    print(name, age)

my_info(age=20, name="Rakib")
```

```python
# 3 Variable length arguments
def average(x, y, z):
    result = (x+y+z)/3
    return result
average(50,50,50)
```

```python
# Infinity Variable length arguments
def average(*num):
    a = 0
    for i in num:
        a = a + i
        result = a/len(num)
        return result
average(5,9)
```
