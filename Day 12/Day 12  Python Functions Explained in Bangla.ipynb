```python
# len function creation use case

l = [1,3,4,22,23,45,67]
def my_len(l):
    if type(l) == list:
        a = 0
        for i in l:
            a = a + 1
        return a
    else:
        return "this is not a list"
```

```python
my_len(l)
```

```python
# to sum print function use case
l2 = [1,2,3,4,"ridoy",6,7,266,"sakib",23,67,"rakib",98,90]
a = 0
for i in l2:
    if type(i) == int:
        a = a + i
print(a)        
```

```python
# to sum defined function creating for use case
def list_sum(x):
    a = 0
    if type(x) == list:
        for i in x:
            if type(i) == int:
                a = a + i
        return a        
```

```python
list_sum(l2)
```

```python
list_sum([23,45,65,78,9])
```
